```python
# level2/방금그곡-올바른 풀이
def solution(m,musicinfos):
    
    def change(txt):
        dic={"C#":"c","D#":"d","E#":"e",'F#':"f","G#":"g","A#":"a"}
        for i in dic:
            if i in txt:
                a=txt.replace(i,dic[i])
                txt=a
        return txt
    music={}
    answer=[]
    for i in musicinfos:
        info=i.split(",")
        sound=change(info[3])
        hour = 1 * (int(info[1].split(':')[0]) - int(info[0].split(':')[0]))
        if hour == 0: 
            time = int(info[1].split(':')[1]) - int(info[0].split(':')[1])
        else: 
            time = 60 * hour + int(info[1].split(':')[1]) - int(info[0].split(':')[1])
        # 재생시간을 sound길이로 나눈 몫과 나머지를 구한 뒤 몫만큼 sound를 곱하고 남은 나머지시간만큼 sound[:r]을 더해준다.
        # music[곡 제목] = [재생된 음, 재생시간]
        q,r=divmod(time,len(sound))
        music[info[2]]=[sound*q+sound[:r],time]
        
    change_m=change(m)
    for i in music:
        if change_m in music[i][0]:
            answer.append([i,music[i][1]])
            
    answer.sort(key=lambda x:-x[1])
    
    if answer:
        return answer[0][0]
    else:
        return "(None)"
    
    
#참고) https://hbj0209.tistory.com/88?category=909677
```

```python
# level2/쿼드압축 후 개수 세기
def solution(arr):
    def check(y,x,n):
        if n==1:
            return [0,1] if arr[y][x]==1 else [1,0]
        left_up=check(y,x,n//2)
        right_up=check(y,x+n//2,n//2)
        left_down=check(y+n//2,x,n//2)
        right_down=check(y+n//2,x+n//2,n//2)
        
        #사분면 네개가 전부 동일한 값일 경우=네 개가 아니라 한 개로 취급한다.
        if left_up==right_up==left_down==right_down==[0,1] or left_up==right_up==left_down==right_down==[1,0]:
            return left_up
        else:
            return list(map(sum,zip(left_up,right_up,left_down,right_down)))
        
    result=check(0,0,len(arr))
    return result

#Review)백준과 같은 문제가 있다.=> 색종이 접기
```

```python
# level2/2개 이하로 다른 비트
def solution(numbers):
    answer = []
    for number in numbers:
        bin_number=list("0"+bin(number)[2:])
        #bin_number이 list니까 join을 통해 string으로 변환을 바로해주는것이다.
        idx="".join(bin_number).rfind("0")
        bin_number[idx]="1"
        
        if number%2==1:
            bin_number[idx+1]="0"
         
        #여기서 int내장함수로 int(s,2)를 해주면 2진수 s를 10진수로 쉽게 변환해준다.
        answer.append(int("".join(bin_number),2))
        
    return answer
```
