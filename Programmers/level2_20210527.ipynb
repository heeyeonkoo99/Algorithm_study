```python
#level1/수포자
def solution(answers):
    scores=[0,0,0]
    answer=[]
    
    # 학생들의 답안 패턴 (규칙적임) 
    pattern1 = range(1, 6)
    pattern2 = [2, 1, 2, 3, 2, 4, 2, 5]
    pattern3 = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5]
    
    #학생들이 정답을 맞출때 스코어 +=1
    for i in range(len(answers)):
        if answer[i]==pattern1[i%len(pattern1)]:
            scores[0]+=1
        if answer[i]==pattern2[i%len(pattern2)]:
            scores[1]+=1
        if answer[i]==pattern3[i%len(pattern3)]:
            scores[2]+=1
    max_score=max(scores)
    for i in range(3):
        if scores[i]==max_score:
            answer.append(i+1)
    return answer
            
```

```python
#level2/위장
def solution(clothes):
    answer={}
    for i in clothes:
        if i[1] in answer:answer[i[1]]+=1
        else: answer[i[1]]=1
    print(answer)
    cnt=1
    for i in answer.values():
        cnt*=(i+1)
    return cnt-1
print(solution([["yellowhat", "headgear"], ["bluesunglasses", "eyewear"], ["green_turban", "headgear"]]))
```

```python
#level2/삼각달팽이
def solution(n):
    answer=[[0 for j in range(1,i+1)] for i in range(1,n+1)] #삼각형 구조 만들기
    x,y=-1,0 #좌표 초기화=>처음 시작은 아래로 내려가기 때문에 x=-1
    num=1
    
    for i in range(n): #방향
        for j in range(i,n): #좌표 구하기
            if i%3==0:
                x+=1
            elif i%3==1:
                y+=1
            else:
                x-=1
                y-=1
            answer[x][y]=num
            num+=1
    return sum(answer,[])
#Review) 진짜 이해하기 어려웠던 알고리즘이다. https://minhamina.tistory.com/58 <=이 블로그 잘 참고하기!
#ㅎㅎ이해했다. 결국 x와 y의 값은 변하지않으니까 그걸 잘 참고해서 j번 그만큼 answer[x][y]=num해주면 되는것이다
```
