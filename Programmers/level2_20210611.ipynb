```python
#level2/메뉴 리뉴얼-1
from itertools import combinations
from collections import Counter

def solution(order,couse):
    answer=[]
    for c in couse:
        temp=[]
        for ords in order:
            combi=combinations(sorted(ords),c)
            temp+=combi
        print("하하")
        counter=Counter(temp)
        print(counter)
        if len(counter)!=0 and max(counter.values())!=1:
            answer+=["".join(f) for f in counter if counter[f]==max(counter.values())]
    return sorted(answer)
print(solution(["ABCFG", "AC", "CDE", "ACDE", "BCFG", "ACDEH"],[2,3,4]))
#Review) counter은 dictionary형태이다.
```

```python
#level2/배달
from collections import deque

def solution(N,road,K):
    nodes={}
    dist={i:float("inf")if i!=1 else 0 for i in range(1,N+1)}
    for v1,v2,d in road:
        nodes[v1]=nodes.get(v1,[])+[(v2,d)]
        nodes[v2]=nodes.get(v2,[])+[(v1,d)]
    que=deque([1])
    
    while que:
        cur_node=que.popleft()
        for nxt_node,d in nodes[cur_node]:
            if dist[nxt_node]>dist[cur_node]+d:
                dist[nxt_node]=dist[cur_node]+d
                que.append(nxt_node)
    return len([True for dist in dist.values() if dist<=K])

#Review)다익스트라 알고리즘/와샬-플로이드 알고리즘으로 최단거리구하기
#다익스트라 알고리즘으로 노드를 탐색하면서 거리를 갱신=>graph/distance/que이렇게 3개는 필수!
```
