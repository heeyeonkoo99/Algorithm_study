```python
#level1/실패율
def solution(N, stages):
    answer = []
    failure=[0 for _ in range(N+1)]
    visited=[False for _ in range(N+1)]
    level=1
    cnt=0
    find=0
    length=len(stages)
    flag=False
    sums=0

    while True:
        for i in range(len(stages)):
            if stages[i]==level:
                cnt+=1
                failure[level]=cnt/length
                stages[i]=0
                
            else:
                continue
        level+=1
        length-=cnt
        cnt=0
        if level==N+1:
            break
    failure=failure[1:]
    visited=visited[1:]
  
    while True:
        
        find=max(failure)
        answer.append(failure.index(find)+1)
        visited[failure.index(find)]=True
        failure[failure.index(find)]=0

        sums=sum(failure)
        if sums==0:
            flag=True
            break
        
        
        
    if flag:
        for i in range(len(visited)):
            if visited[i]==False:
                answer.append(i+1)
                visited[i]=True
            if len(answer)==N:
                break

       

    return answer
print(solution(5,[2, 1, 2, 6, 2, 4, 3, 3]))
#Review)후아..시간짱많이 걸렸지만 쾌감은 ㅋㅋㅋㅋ굿굿이다. 암튼 좀더 직관적이고 바로 문제를 제대로 이해하는 연습이 필요하겠다!
```
