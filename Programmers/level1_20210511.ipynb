```python
#level1/크레인 인형뽑기 게임-1
def solution(board,moves):
    stack=[]
    count=0
    for m in moves:
        for i in range(len(board)):
            if board[i][m-1]!=0:
                if stack and stack[-1]==board[i][m-1]:
                    stack.pop()
                    count+=2
                else:
                    stack.append(board[i][m-1])
                board[i][m-1]=0
                break
    return count
print(solution([[0,0,0,0,0],[0,0,1,0,3],[0,2,5,0,1],[4,2,4,4,2],[3,5,1,3,1]],[1,5,3,5,1,2,1,4]))
```

```python
#level1/크레인 인형뽑기 게임-2
from collections import deque
def solution(board,moves):
    answer=0
    board_stack=[deque() for _ in range(len(board))]
    bucket=deque()
    
    for row in board[::-1]:
        for i in range(len(row)):
            if not row[i]==0:
                board_stack[i].append(row[i])
    for num in moves:
        if board_stack[num-1]:
            doll=board_stack[num-1].pop()
            
            if bucket:
                if doll==bucket[-1]:
                    bucket.pop()
                    answer+=2
                else:
                    bucket.append(doll)
            else:
                bucket.append(doll)
    return answer

print(solution([[0,0,0,0,0],[0,0,1,0,3],[0,2,5,0,1],[4,2,4,4,2],[3,5,1,3,1]],[1,5,3,5,1,2,1,4]))   
```

```python
#level1/체육복
def solution(n,lost,reserve):
    new_lost=set(lost)-set(reserve)
    new_reserve=set(reserve)-set(lost)
    for i in new_lost:
        if i+1 in new_reserve:
            new_reserve.remove(i+1)
        elif i-1 in new_reserve:
            new_reserve.remove(i-1)
        else:
            n-=1
    return n
```
